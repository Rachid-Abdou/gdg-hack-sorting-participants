```python
def get_github_users_data(urls, data):
    # Create an empty list to store user data
    user_data_list = []
    
    for url in urls:
        # GitHub API endpoint for retrieving user information
        # url = f'https://api.github.com/users/{username}'
        
        # Send a GET request to the endpoint
        response = requests.get(url)
        
        # Check if the request was successful
        if response.status_code == 200:
            # Convert the response JSON to a Python dictionary
            user_data = response.json()
            
            # Get user's repositories
            repos_url = user_data['repos_url']
            repos_response = requests.get(repos_url)
            
            if repos_response.status_code == 200:
                # Convert the response JSON to a list of repositories
                repositories = repos_response.json()
                
                # Calculate total stars and forks for all repositories
                total_stars = sum(repo['stargazers_count'] for repo in repositories)
                total_forks = sum(repo['forks_count'] for repo in repositories)
                
                # Add user data to the list
                user_info = {
                    "github" : user_data['url'],
                    'Username': user_data['login'],
                    'Public Repositories': user_data['public_repos'],
                    'Followers': user_data['followers'],
                    'Total Stars': total_stars,
                    'Total Forks': total_forks
                }
                user_data_list.append(user_info)
            else:
                print(f"Failed to retrieve repositories data for {url}. Status code: {repos_response.status_code}")
        else:
            print(f"Failed to retrieve user data for {url}. Status code: {response.status_code}")
    
    try: # Convert the list of user data into a DataFrame
        df = pd.DataFrame(user_data_list)
        df = df.merge(data)
        print(f'GitHub users data has been exported to github_users_data.csv')
        return df

    except:
        return None
```

```python
import time 

def eval_git_participerte(data):
    df = data
    # Calculate the components
    eval_git_public_repositories = df['Public Repositories'] / 5
    eval_git_followers = df['Followers'] / 5
    eval_git_total_stars = df['Total Stars'] / 5

    # Apply the min and sum operations to get the final 'eval-git' column
    df['eval-git'] = (eval_git_public_repositories.clip(upper=20) +
                    eval_git_followers.clip(upper=10) +
                    eval_git_total_stars.clip(upper=10))
    return df

# Function to evaluate motivation letter using GPT API
def eval_motivation(data):
    df = data
    eval_mot_responses = []
    for motivation_text in df['motivation']:
        
        prompt = f" We are GDG Algiers, the foremost scientific club in Algeria. Currently, we're gearing up for our upcoming event, DevFest, which stands as the premier hackathon in Algeria, boasting 30 teams dedicated to crafting innovative technological solutions. Your role is to evaluate the motivation letters submitted by participants. You will be provided with a motivation letter, and your task is to assign a score between 0 and 20, with 20 being the highest (the answers must be just the score as intiger, and do not write with it any character). \nMotivation Letter: {motivation_text}"
        response = llm.invoke(prompt)

        # time.sleep(30)
        eval_mot_responses.append(response.content)

    df['eval_motivation'] = eval_mot_responses

    return df


def eval_project(data):
    df = data
    eval_projects_responses = []
    for project_text in df['projects']:
        
        prompt_projects = f"We are GDG Algiers, the foremost scientific club in Algeria. Currently, we're gearing up for our upcoming event, DevFest, which stands as the premier hackathon in Algeria, boasting 30 teams dedicated to crafting innovative technological solutions. Your role is to evaluate the projects  \nthat the participate worked on in before. You will be provided with a text contain projects, and your task is to assign a score between 0 and 10, with 10 being the highest. We give [0 to 10] points for each project, and if he has a good explanation for the project he will get more points. you must be hard in the evaluation of the proejcts (give me just the result of the sum of all the projects score(just one number). the answers must be just the total score as intiger and not write with it any character). \nMotivation Letter: {project_text}"
        response = llm.invoke(prompt_projects)

        # time.sleep(30)
        eval_projects_responses.append(response.content)

    df['eval_projects'] = eval_projects_responses

    return df
```

```python
import re 
# Function to extract integer from a string
def extract_integer(value):
    if isinstance(value, str):
        integers = re.findall(r'\b\d+\b', value)
        if integers:
            return int(integers[0])
    return value
```

```python
import requests
import pandas as pd

# Replace 'username' with the GitHub username you want to retrieve data for
username = 'wadjih-bencheikh18'

# GitHub API endpoint for retrieving user information
url = f'https://api.github.com/users/{username}'

# Send a GET request to the endpoint
response = requests.get(url)

# Check if the request was successful
if response.status_code == 200:
    # Convert the response JSON to a Python dictionary
    user_data = response.json()
    
    # Get user's repositories
    repos_url = user_data['repos_url']
    repos_response = requests.get(repos_url)
    
    if repos_response.status_code == 200:
        # Convert the response JSON to a list of repositories
        repositories = repos_response.json()
        
        # Create a list to hold repository information
        repo_list = []
        for repo in repositories:
            repo_info = {
                'Repository Name': repo['name'],
                'Description': repo['description'],
                'Language': repo['language'],
                'Stars': repo['stargazers_count'],
                'Forks': repo['forks_count']
            }
            repo_list.append(repo_info)
        
        # Convert the list of repositories into a DataFrame
        df = pd.DataFrame(repo_list)
        
        # Save the DataFrame to a CSV file
        df.to_csv(f'{username}_repositories.csv', index=False)
        print(f'Repository data for {username} has been exported to {username}_repositories.csv')
    else:
        print(f"Failed to retrieve repositories data. Status code: {repos_response.status_code}")
else:
    print(f"Failed to retrieve user data. Status code: {response.status_code}")
```

```python
import requests
import pandas as pd

def get_github_users_data(usernames, output_csv_path='github_users_data.csv'):
    # Create an empty list to store user data
    user_data_list = []
    
    for username in usernames:
        # GitHub API endpoint for retrieving user information
        url = f'https://api.github.com/users/{username}'
        
        # Send a GET request to the endpoint
        response = requests.get(url)
        
        # Check if the request was successful
        if response.status_code == 200:
            # Convert the response JSON to a Python dictionary
            user_data = response.json()
            
            # Get user's repositories
            repos_url = user_data['repos_url']
            repos_response = requests.get(repos_url)
            
            if repos_response.status_code == 200:
                # Convert the response JSON to a list of repositories
                repositories = repos_response.json()
                
                # Calculate total stars and forks for all repositories
                total_stars = sum(repo['stargazers_count'] for repo in repositories)
                total_forks = sum(repo['forks_count'] for repo in repositories)
                
                # Add user data to the list
                user_info = {
                    "github" : user_data['url'],
                    'Username': user_data['login'],
                    # 'Name': user_data['name'],
                    # 'Bio': user_data['bio'],
                    'Public Repositories': user_data['public_repos'],
                    'Followers': user_data['followers'],
                    # 'Following': user_data['following'],
                    'Total Stars': total_stars,
                    'Total Forks': total_forks
                }
                user_data_list.append(user_info)
            else:
                print(f"Failed to retrieve repositories data for {username}. Status code: {repos_response.status_code}")
        else:
            print(f"Failed to retrieve user data for {username}. Status code: {response.status_code}")
    
    # Convert the list of user data into a DataFrame
    df = pd.DataFrame(user_data_list)
    
    # Save the DataFrame to a CSV file
    df.to_csv(output_csv_path, index=False)
    print(f'GitHub users data has been exported to github_users_data.csv')

```

```python
# Example usage:
usernames = ['AssoulYasser', '3b1b', 'Fellah-wassim', 'BoumedineBillal', 'wadjih-bencheikh18']  # Replace with the list of GitHub usernames
get_github_users_data(usernames)
```
